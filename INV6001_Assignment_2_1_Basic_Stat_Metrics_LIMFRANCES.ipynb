###Co-variance

Dispersion metric - average distances between the values and the mean 

Given attributes are x and y.
Co-variance is the variance for more than one attribute, like x and y. 
While variance is just for x.

```python
#return unique values for GearType in cars dataframe

pd.unique(cars.GearType)
```
